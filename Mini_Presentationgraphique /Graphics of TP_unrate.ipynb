## Présentation du dataset

Exploration du taux de chômage mensuel c'est à dire le taux de personne en age de travail et non pas de boulet ou job. C'est un réel indicateur des communes du pays. 

On travaillera sur le fichier "unrate.csv" qui présente le taux de chomage au UEcentre entre Janvier 1948 et janvier 2018. 

```python
import pandas as pd
unrate = pd.read_csv("unrate.csv")
unrate.head()
```

```python
### Convertir la colonne Date à des vraies dates au lieu des chaînes de caractère 
unrate["DATE"]= pd.to_datetime(unrate["DATE"])
unrate.head(12)
```

```python

import matplotlib.pyplot as plt
```

```python
% matplotlib inline
```

```python
## Visualiser le taux de chomâge en 1948
first_year = unrate[0:12]
plt.plot(first_year["DATE"], first_year["UNRATE"])
plt.xticks(rotation=90)
plt.xlabel("Mois ")
plt.ylabel("Taux de chômage (en %)")
plt.title("Evolution du taux de chômage en 1948 aux USA")
plt.show()
```

```python
### le taux de chomâge 1948 et 1949
## 1e 12 lignes de l'année 1948 
first_year = unrate[0:12]
first_year
```

```python
## 2ème  12 lignes de l'année 1944
second_year = unrate[12:24]
second_year
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(first_year["DATE"],first_year["UNRATE"])
ax2.plot(second_year["DATE"],second_year["UNRATE"])
plt.show()
```

```python
### ajuster la taille d'affichage des graphiques
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(first_year["DATE"],first_year["UNRATE"])
ax1.set_title("Taux de chômage en 1948")

ax2.plot(second_year["DATE"],second_year["UNRATE"])
ax2.set_title("Taux de chômage en 1949")
plt.show()
```

```python
## afficher plusieurs courbes à la fois del'année 1948 à 1952
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'],subset['UNRATE'])

plt.show()
```

```python
### extraire les valeurs du mois de la colonne date
unrate["MONTH"]= unrate['DATE'].dt.month

## génèrer les courbes des années 1948 et 1949 
fig = plt.figure(figsize=(6,3))
plt.plot(unrate[0:12]["MONTH"],unrate[0:12]["UNRATE"],c = "red")
plt.plot(unrate[12:24]["MONTH"],unrate[12:24]["UNRATE"],c = "blue")
plt.show()
```

```python
## génèrer les graphiques des années 1948 à 1952
fig = plt.figure(figsize=(8,5))
colors=['red',"blue",'green','orange','black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset["MONTH"],subset["UNRATE"], c=colors[i])
plt.show()
```

```python
## rajouter la légende pour chaque année 1948 à 1952
fig = plt.figure(figsize=(8,5))
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset["MONTH"],subset["UNRATE"], c=colors[i], label=label)
plt.legend(loc="upper left")
plt.show()
```

```python
## courbe finalisée
## rajouter le titre du graphique et les labels des axes x et y 
fig = plt.figure(figsize=(8,5))
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset["MONTH"],subset["UNRATE"], c=colors[i], label=label)
plt.title("Evolution du taux de chômage de 1948 à 1952")
plt.xlabel('Mois(de 1 à 12)')
plt.ylabel('Taux de chômage(en %)')
plt.legend(loc="upper left")
plt.show()
```
