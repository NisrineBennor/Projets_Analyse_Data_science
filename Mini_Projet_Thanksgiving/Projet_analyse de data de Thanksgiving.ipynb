### Convertir l'âge en une valeur numérique

```python
data["Age"].value_counts()
```

```python
## Convertir l'age d'une chaine de caractère à une valeur numarique en gardant que la première partie
def extract_age(str_age):
    if pd.isnull(str_age):
        return None 
    str_age = str_age.split(" ")[0]
    str_age = str_age.replace("+","")
    return int(str_age)

data["int_age"]= data['Age'].apply(extract_age)
data["int_age"].describe()
```

### Convertir les revenus en valeurs numériques

```python
data["How much total combined money did all members of your HOUSEHOLD earn last year?"]
```

```python
### Convertir les revenus en valeurs numériques
def extract_income(income_str):
    if pd.isnull(income_str):
        return None 
    income_str = income_str.split(" ")[0]
    if income_str == 'Prefer':
        return None 
    income_str = income_str.replace("$","")
    income_str = income_str.replace(",","")
    return int(income_str)

data["int_income"]=data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(extract_income)
data["int_income"].describe()
```

### Corrélation entre distance et revenus

```python
### de quelle manière les personnes gagnant moins de 150000 voyagent???
less_150000=data[data["int_income"]<150000]
```

```python
less_150000["How far will you travel for Thanksgiving?"].value_counts()
```

```python
### de quelle manière les personnes gagnant plus de 150000 voyagent???
more_150000=data[data["int_income"]>150000]
```

```python
more_150000["How far will you travel for Thanksgiving?"].value_counts()
```

### Lien entre passer le Thanksgiving entre amis avec l'âge et le revenu 

```python
### Cas d'âge: moyenne d'age des sondés pour chaque catégorie de questions 
import numpy as np
data.pivot_table(index="Have you ever tried to meet up with hometown friends on Thanksgiving night?",columns='Have you ever attended a "Friendsgiving?"',values="int_age", aggfunc=np.mean)
```

```python
## Cas de revenu: moyenne de revenues des sondés pour chaque catégorie de questions 
data.pivot_table(index="Have you ever tried to meet up with hometown friends on Thanksgiving night?",columns='Have you ever attended a "Friendsgiving?"',values="int_income", aggfunc=np.mean)
```

On peut dire que les gens qui sont plus jeunes ont tendance à fêterle Thanksgiving avec leur que amis et même pour le cas des revenus; les gens qui gagnent moins fêtent le thankksgiving avec leur amis. 
- peut être qu'il existe une hypothése entre le jeune age et le niveau de revenus.
