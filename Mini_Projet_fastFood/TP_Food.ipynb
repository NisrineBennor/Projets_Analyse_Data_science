```python
## manipuler 2 colonnes 
gram_of_protein_per_gram_of_water = food_info["Protein_(g)"]/food_info["Water_(g)"]
print(gram_of_protein_per_gram_of_water[0:5])
```

```python
milligrams_of_calcium_and_iron = food_info["Calcium_(mg)"]+food_info["Iron_(mg)"]
print(milligrams_of_calcium_and_iron[0:5])
```

```python
## Créer un indice nutritionnel avec la formule suivante:
##  Score = 2 * Protein - 0.75 * Lipid
Protein_2 = food_info["Protein_(g)"]*2
Lipid = food_info["Lipid_Tot_(g)"]*(-0.75)
rating = Protein_2 + Lipid
print(rating[0:5])
```

```python
## normaliser la valiable Protein
max_protein = food_info["Protein_(g)"].max()
normalized_protein= food_info["Protein_(g)"]/max_protein
print(normalized_protein[0:5])
```

```python
## normaliser la valiable Lipid
max_lipid = food_info["Lipid_Tot_(g)"].max()
normalized_Lipid= food_info["Lipid_Tot_(g)"]/max_lipid
print(normalized_Lipid[0:5])
```

```python
## ajouter 2 colonnes les variables normalisées
food_info["normalized_protein"]=normalized_protein
food_info["normalized_Lipid"]=normalized_Lipid
food_info.head(3)
```

```python
## Créer un indice nutritionnel normalisé
Norm_prot = food_info["normalized_protein"]*2
Norm_Lip = food_info["normalized_Lipid"]*(-0.75)
Norm_Nutr_Index = Norm_prot+ Norm_Lip
food_info["Norm_Nutr_Index"]= Norm_Nutr_Index
food_info
```

```python
## Trier l'indice nutrionnel de plus grand au plus petit.
food_info.sort_values("Norm_Nutr_Index", inplace=True, ascending=False)
food_info
```
